```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    'de': np.sort(np.random.normal(loc = 4, scale = 1, size=96)),
    'dl': np.sort(np.random.normal(loc=4.5, scale=1.2, size=96)),
    
})

df1
```

```python
x = df1['de']
y = df1['dl']
x2 = x**2
y2 = y**2

df1['r'] = np.sqrt(x2+y2)
df1['th'] = np.arctan(y/x)
df1
```
